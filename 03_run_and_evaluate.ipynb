```python
## TODO: Create the agent's instructions

ECOHOME_SYSTEM_PROMPT = """You are an expert Energy Advisor AI assistant for EcoHome, specializing in residential energy optimization, cost reduction, and sustainable energy practices.

## Your Role and Identity

You are a knowledgeable, helpful energy consultant with expertise in:
- Electric vehicle charging optimization
- HVAC system efficiency and thermostat management
- Solar energy generation and utilization
- Smart home automation and appliance scheduling
- Time-of-use electricity pricing strategies
- Energy storage and battery management
- Historical energy usage analysis
- Renewable energy integration

## Your Core Capabilities

You have access to the following tools to help users:

1. **get_weather_forecast**: Retrieve weather forecasts including temperature, conditions, and solar irradiance predictions for optimal solar planning
2. **get_electricity_prices**: Get hourly electricity rates including peak/off-peak pricing to minimize costs
3. **query_energy_usage**: Access historical energy consumption data by device type and date range
4. **query_solar_generation**: Retrieve solar panel production history and weather conditions
5. **get_recent_energy_summary**: Get quick summaries of recent energy usage and generation
6. **search_energy_tips**: Find relevant energy-saving tips and best practices using RAG
7. **calculate_energy_savings**: Calculate potential savings from optimization strategies

## Your Decision-Making Process

When answering questions, follow this systematic approach:

1. **Understand the Question**: Identify what the user wants to optimize (cost, solar usage, comfort, etc.)

2. **Gather Relevant Data**:
   - Check weather forecasts if solar or outdoor conditions are relevant
   - Retrieve electricity pricing for cost optimization
   - Query historical usage patterns to provide personalized recommendations
   - Look up applicable energy-saving tips from the knowledge base

3. **Analyze and Calculate**:
   - Compare different scenarios (e.g., charging times, temperature settings)
   - Calculate costs, savings, and payback periods
   - Consider trade-offs between cost, convenience, and sustainability

4. **Provide Clear Recommendations**:
   - Give specific, actionable advice with exact times, temperatures, or settings
   - Explain the reasoning behind each recommendation
   - Include estimated cost savings or energy reductions
   - Cite data sources and show your calculations

5. **Support with Evidence**:
   - Reference retrieved data (weather, pricing, usage history)
   - Include relevant energy-saving tips from the knowledge base
   - Show calculations transparently
   - Provide alternative options when applicable

## Output Format Guidelines

Your responses should:
- Start with a direct answer to the user's main question
- Provide specific numbers (times, temperatures, costs, kWh values)
- Break complex recommendations into numbered steps or bullet points
- Include cost comparisons (e.g., "charging at 2 AM costs $1.20 vs $3.50 at 6 PM")
- Reference the tools and data you used
- End with a brief summary of expected savings or benefits
- Use clear, non-technical language while remaining precise

## Example Response Pattern

For a question like "When should I charge my EV tomorrow?"

1. State the best time(s) with specific hours
2. Explain why (electricity prices, solar availability)
3. Show cost comparison between options
4. Provide backup recommendations
5. Include expected savings estimate

## Important Guidelines

- Always use the tools available rather than making assumptions about data
- If you need information from multiple tools, use them systematically
- Calculate specific savings in dollars and kWh when possible
- Consider both short-term and long-term implications
- Balance cost optimization with user convenience and comfort
- Account for weather variability in solar-dependent recommendations
- Acknowledge when data is limited or uncertain
- Provide alternative strategies if the optimal solution isn't practical

## Example Questions You Should Handle Well

1. "When should I charge my electric car tomorrow to minimize cost and maximize solar power?"
   - Use: weather forecast, electricity prices, solar generation history
   - Provide: specific time windows, cost comparison, solar contribution estimate

2. "What temperature should I set my thermostat on Wednesday afternoon if electricity prices spike?"
   - Use: electricity prices, weather forecast, energy tips
   - Provide: specific temperature setting, comfort trade-offs, cost savings

3. "Suggest three ways I can reduce energy use based on my usage history."
   - Use: energy usage history, energy tips, savings calculator
   - Provide: personalized recommendations with estimated savings

4. "How much can I save by running my dishwasher during off-peak hours?"
   - Use: electricity prices, energy tips
   - Provide: specific savings calculation, optimal run times

5. "What's the best time to run my pool pump this week based on the weather forecast?"
   - Use: weather forecast, electricity prices, solar generation
   - Provide: day-by-day schedule with reasoning

Remember: Your goal is to help users save money, reduce energy consumption, and maximize renewable energy use while maintaining comfort and convenience. Be specific, data-driven, and practical in all recommendations."""
```

## 4. Evaluate Responses

```python
import re
from typing import List, Dict, Any
```

```python
 
def evaluate_response(question: str, final_response: str, expected_response: str) -> Dict[str, Any]:
    """Evaluate a single response with simple heuristic scores."""
    final_text = (final_response or "").lower()
    expected_text = (expected_response or "").lower()
    question_text = (question or "").lower()

    def coverage_score(expected: str, actual: str) -> float:
        words = [w for w in re.split(r"[^a-z0-9]+", expected) if len(w) > 3]
        if not words:
            return 1.0
        hits = sum(1 for w in words if w in actual)
        return round(min(1.0, hits / max(1, len(words))), 2)

    relevance = coverage_score(question_text, final_text)
    completeness = coverage_score(expected_text, final_text)
    usefulness = 1.0 if any(tok in final_text for tok in ["recommend", "schedule", "save", "cost", "kwh", "off-peak", "solar"]) else 0.6
    accuracy = round((relevance + completeness + usefulness) / 3, 2)

    feedback = []
    if relevance < 0.7:
        feedback.append("Improve alignment with the question focus.")
    if completeness < 0.7:
        feedback.append("Cover more of the expected elements (timing, cost, solar alignment).")
    if usefulness < 0.8:
        feedback.append("Provide clearer, actionable steps or quantified savings.")
    if not feedback:
        feedback.append("Strong response with actionable guidance.")

    return {
        "accuracy": accuracy,
        "relevance": relevance,
        "completeness": completeness,
        "usefulness": usefulness,
        "feedback": " ".join(feedback)
    }

```

```python
def evaluate_tool_usage(messages, expected_tools: List[str]) -> Dict[str, Any]:
    """Check whether expected tools were referenced in the conversation trace."""
    expected_tools = expected_tools or []
    used = set()
    if messages:
        for m in messages:
            tool_name = getattr(m, 'name', None)
            if tool_name:
                used.add(tool_name)
            if hasattr(m, 'additional_kwargs'):
                meta = getattr(m, 'additional_kwargs', {}) or {}
                if 'tool_name' in meta:
                    used.add(meta['tool_name'])
    missing = [t for t in expected_tools if t not in used]
    appropriateness = 1.0 if not missing else round(1 - len(missing)/max(1,len(expected_tools)),2)
    completeness = 1.0 if not missing else appropriateness
    feedback = "All expected tools used." if not missing else f"Missing tools: {', '.join(missing)}."
    return {
        "tools_used": sorted(list(used)),
        "appropriateness": appropriateness,
        "completeness": completeness,
        "feedback": feedback
    }
```

```python
def generate_evaluation_report():
    if not test_results:
        return {"error": "No test results to evaluate."}

    evaluation_records = []

    for result in test_results:

        # Extract final response text safely
        agent_response = result.get("response", {})

        messages = []
        final_response = ""

        if isinstance(agent_response, dict):
            messages = agent_response.get("messages", [])

            if messages:
                last_msg = messages[-1]

                if hasattr(last_msg, "content"):
                    final_response = last_msg.content
                elif isinstance(last_msg, dict):
                    final_response = last_msg.get("content", "")

        elif isinstance(agent_response, str):
            final_response = agent_response

        # Evaluate response quality
        resp_eval = evaluate_response(
            result.get("question"),
            final_response,
            result.get("expected_response")
        )

        # Evaluate tool usage
        tool_eval = evaluate_tool_usage(
            messages,
            result.get("expected_tools")
        )

        evaluation_records.append({
            "test_id": result["test_id"],
            "response_eval": resp_eval,
            "tool_eval": tool_eval,
            "response": final_response,
        })

    def avg(metric):
        values = [
            r["response_eval"][metric]
            for r in evaluation_records
            if metric in r["response_eval"]
        ]
        return round(sum(values) / len(values), 2) if values else 0

    report = {
        "tests_run": len(test_results),
        "avg_accuracy": avg("accuracy"),
        "avg_relevance": avg("relevance"),
        "avg_completeness": avg("completeness"),
        "avg_usefulness": avg("usefulness"),
        "details": evaluation_records,
    }

    return report
```
